```python
# Assuming dataset['alcohol'] is your data column
data = dataset['alcohol']

# Creating the histogram with different colors for each bar
n, bins, patches = plt.hist(data, bins=15, color='skyblue', alpha=0.7)

# Adding labels and title
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')

# Adding text annotations for each bar
for count, bin_edge, patch in zip(n, bins[:-1], patches):
    plt.text(bin_edge + 0.2, count, f'{bin_edge:.2f}', ha='center', va='bottom', color='black', fontsize=7)

# Displaying the plot
plt.show()
```

# REPORT
## From the histogram, it can be seen that alcohol content 9.27 has the highest frequency and 14.03 has the lowest 

```python
plt.scatter(dataset['alcohol'], dataset['pH'])
plt.title('Scatter Plot: Alcohol Content vs pH')
plt.xlabel('Alcohol Content')
plt.ylabel('pH')
plt.show()
```

# Report
### It can be seen that wine within the 3.2 and 3.4 region has a very dense population

```python
plt.bar(dataset['quality'].value_counts().index, dataset['quality'].value_counts().values)
plt.title('Distribution of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

# Report
### It can be seen that most wines in the dataset have the quality of 5 and least wine has a quality of 3
